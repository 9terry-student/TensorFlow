# 6мһҘ. лЎңм§ҖмҠӨнӢұ нҡҢк·Җ лӘЁлҚё: м°ё кұ°м§“ нҢҗлӢЁн•ҳкё°

## 5. н…җм„ңн”ҢлЎңмҡ°м—җм„ң мӢӨн–үн•ҳлҠ” лЎңм§ҖмҠӨнӢұ нҡҢк·Җ лӘЁлҚё

### 1. нҷҳкІҪ мӨҖл№„

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2. лҚ°мқҙн„° мӨҖл№„

```python
x=np.array([2,4,6,8,10,12,14])
y=np.array([0,0,0,1,1,1,1])
```

### 3. лӘЁлҚё мӢӨн–ү

```python
model=Sequential()
model.add(Dense(1,input_dim=1,activation='sigmoid'))

# көҗм°Ё м—”нҠёлЎңн”ј мҳӨм°Ё н•ЁмҲҳлҘј мқҙмҡ©н•ҳкё° мң„н•ҙ 'binary_crossentropy'лЎң м„Өм •
model.compile(optimizer='sgd',loss='binary_crossentropy')
model.fit(x,y,epochs=1000)
```

### 4. лӘЁлҚё н…ҢмҠӨнҠё

```python
# к·ёлһҳн”„лЎң нҷ•мқё

plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')
plt.show()
```

```python
import tensorflow as tf

# мһ„мқҳмқҳ н•ҷмҠө мӢңк°„мқ„ м§‘м–ҙл„Јм–ҙ н•©кІ© мҳҲмғҒ нҷ•лҘ  мҳҲмёЎ

hour=7

input_data=tf.constant([[hour]],dtype=tf.float32)
prediction=model.predict(input_data)[0][0]

print('%.fмӢңк°„мқ„ кіөл¶Җн•  кІҪмҡ°, н•©кІ© мҳҲмғҒ нҷ•лҘ мқҖ %.01f%%мһ…лӢҲлӢӨ' %(hour,prediction*100))
```
